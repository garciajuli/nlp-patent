```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv (r'fulltext-EP3500000.txt',sep = "\t", encoding = "utf-8",nrows=1000, header = None)
data.columns = ['_', 'publicationNumber', 'publicationKind', 'date', 'language', 'textType', 'unicityKey', 'content']
new_data = data[['publicationNumber', 'publicationKind', 'date', 'language', 'textType', 'unicityKey', 'content']]
```

```python
new_data['title'] = new_data[new_data['textType'] == 'TITLE']['content']
new_data['abstr'] = new_data[new_data['textType'] == 'ABSTR']['content']
new_data['descr'] = new_data[new_data['textType'] == 'DESCR']['content']
new_data['claim'] = new_data[new_data['textType'] == 'CLAIM']['content']
new_data['amend'] = new_data[new_data['textType'] == 'AMEND']['content']
new_data['acstm'] = new_data[new_data['textType'] == 'ACSTM']['content']
new_data['srprt'] = new_data[new_data['textType'] == 'SRPRT']['content']
new_data['pdfep'] = new_data[new_data['textType'] == 'PDFEP']['content']
```

```python
new_data
```

```python
def list_reduce(L):
    if type(L) != list:
        return(L)
    L = [x for x in L if str(x) != 'nan']
    if all(elem == 'nan' for elem in L):
        return(np.nan)
    if all(elem == L[0] for elem in L):
        return(L[0])
    else:
        return(L)
```

```python
test = new_data.groupby(['publicationNumber','language']).agg(lambda x: x.tolist())
```

```python
L = ['nan', 'nan', 'nan']
L = list_reduce(L)
L
```

```python
for column in test.columns:
    test[column] = test[column].transform(list_reduce)
test = test[['date', 'title', 'abstr', 'descr', 'claim','amend','acstm']]
```
